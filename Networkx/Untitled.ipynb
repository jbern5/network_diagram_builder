```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
df = pd.read_csv("network_data.csv")
```

```python
for index, row in df.iterrows():
    G.add_node(row['cati'])
    for edge in list(row['connections']):
        print(edge)
        G.add_edge(row['cati'], edge)
```

```python
nx.draw(G, with_labels=True)
plt.draw()
plt.show()
```
